## Linear Regression (Just to understand Concept)

```python
#Sample heights in centimeters 
X = [[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]]

#Corresponding weights in Kilograms
y = [63,81,56,91,47,57,76,72,62,48]

#Import Linear Regression Class from scikit-Learn
from sklearn.linear_model import LinearRegression

#Create Linear Regression object
regressor = LinearRegression()

#Train the algorithm with given data (i/p and o/p)
regressor.fit(X,y)

new = [[175]]

pred = regressor.predict(new)

print('Weight of the person will be',pred[0],'kgs')

print("Accuracy:",regressor.score(X,y)*100)

print(regressor.coef_)
print(regressor.intercept_)

pred = 0.67461045 * 175 -38.45508707607701
print(pred)

y

regressor.predict(X)

import matplotlib.pyplot as plt
plt.xlabel('Height (cms)')
plt.ylabel('Weight (kgs)')
plt.scatter(X,y)
plt.plot(X,regressor.predict(X),color='red')
```

```python
A = [[1],[2],[3],[4],[5]]
b = [[100],[200],[300],[400],[500]]
regressor.fit(A,b)

regressor.score(A,b)*100

regressor.predict([[23]])

plt.xlabel('Height (cms)')
plt.ylabel('Weight (kgs)')
plt.scatter(A,b)
plt.plot(A,regressor.predict(A),color='red')
```

## Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('dataset/Salary.csv')  # salary.csv 

print('Data Size:',data.shape)

print('Columns:', data.columns)

X = data['YearsExperience'].values
y = data['Salary']

X = X.reshape(-1,1)

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)

len(X)
len(X_train)
len(X_test)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(X_train,y_train)
regressor.score(X_train,y_train)*100
regressor.score(X_test,y_test)*100
y_pred = regressor.predict(X_test)
result = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
print(result)

regressor.predict([[15]])[0]

from sklearn import metrics
print('MSE:',metrics.mean_squared_error(y_test,y_pred))

from sklearn.metrics import mean_squared_error
#Given Values
Y_true = [1,1,2,2,4] #Y_true = Y (original values)
#Calculated values
Y_pred = [0.6,1.29,1.99,2.69,3.4] # Y_pred = Y

#Calculation of mean squared error(MSE)
mean_squared_error(Y_true,Y_pred)

x=(1-0.6)**2+(1-1.29)**2+(2-1.99)**2+(2-2.69)**2+(4-3.4)**2
x/5

print('MAE:',metrics.mean_absolute_error(y_test,y_pred))

#Avg(Abs(Actual-Forecast)/Abs(Actual))*100
np.mean(np.abs(y_test-y_pred)/np.abs(y_test)) * 100

regressor.score(X_test,y_test)*100
```

## Polynomial Regression

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset/Salary.csv')
```

```python
X = data['YearsExperience'].values
X = X.reshape(-1,1)
y = data['salary']
```

```python
from  sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
print('Accuracy:',np.round(regressor.score(X, y)*100,3))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
Xp = poly.fit_transform(X)
regressor.fit(Xp,y)
```

```python
print('Accuracy:', np.round(regressor.score(Xp, y)*100,3))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, regressor.predict(Xp), color='r')
```

```python
acc = []
for n in range(2,31):
    pf = PolynomialFeatures(degree=n)
    Xp = pf.fit_transform(X)
    regressor.fit(Xp, y)
    acc.append(np.round(regressor.score(Xp, y)*100,3))
    
```

```python
import matplotlib.pyplot as plt
max(acc)
```

```python
plt.xlabel('Degree')
plt.ylabel('Accuracy')
plt.xlim(0,30)
plt.ylim(80,100)
plt.grid()
plt.plot(range(2,31), acc, color='red')
plt.scatter(acc.index(max(acc))+2,max(acc),color='blue',s=50)
```

```python
pf = PolynomialFeatures(degree=13)
Xp = pf.fit_transform(X)
regressor.fit(Xp, y)
```

```python
regressor.score(Xp, y) * 100
```
