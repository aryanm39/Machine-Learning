```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("housearea.csv")
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
import math as mt
bmean=mt.floor(df['bedroom'].mean())
df.bedroom=df.bedroom.fillna(bmean)
df
```

```python
x=df.drop('price',axis='columns')
x
```

```python
y=df['price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict([[4200,4,5]])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("crash.csv")
df.head()
```

```python
x=df.drop('fate',axis='columns')
x
```

```python
y=df['fate']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[25,110]])
```

```python
model.predict([[38,140]])
```

```python
import pandas as pd
df=pd.read_csv("position_sal.csv")
df.head()
```

```python
x=df.iloc[:,1:2].values
y=df.iloc[:,2:].values
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
reg
```

```python
reg.predict([[11]])
```

```python
reg.predict([[12]])
```

```python
from sklearn.metrics import mean_squared_error
psal=reg.predict(x)
mse=mean_squared_error(y,psal)
print(mse)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
reg.predict(poly.fit_transform([[11]]))
```

```python
reg.predict(poly.fit_transform([[12]]))
```

```python
from sklearn.metrics import mean_squared_error
y_pred=reg.predict(x_poly)
mse=mean_squared_error(y_pred,y)
print(mse)
```
