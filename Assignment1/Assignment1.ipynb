## Q1

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns

iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

df['target'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
sns.scatterplot(data = df , x = "sepal length (cm)", y  = "sepal width (cm)", hue='target', style="target")
plt.legend(title="Species")
```

## Q4

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
 

iris = load_iris()
X = iris.data          
y = iris.target        
 

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
 
model = LogisticRegression()
model.fit(X_train, y_train)

new_flower = [[5.1, 3.5, 1.4, 0.2]]              
new_flower_pca = pca.transform(new_flower)       
prediction = model.predict(new_flower_pca)

print("Predicted class:", iris.target_names[prediction])
```
